```python
##

class Table:
    def __init__(self, tableNo, waiterName, status):
        self.tn = tableNo
        self.wn = waiterName
        self.st = status
        
def findWaiterWiseTotalNoOdTables(t_list):
    w_list = {}
    for i in t_list:
        if i.wn in w_list:
            w_list[i.wn] += 1
        else:
            w_list[i.wn] = 1
    
    return w_list

def findWaiterNameByTableNo(t_list, tb_no):
    for i in t_list:
        if i.tn == tb_no:
            return i.wn
    
    return None   

if __name__ == "__main__":
    N = int(input())
    t_list = []
    for i in range(N):
        tno = int(input())
        wname = input()
        status = input()
        t_list.append(Table(tno, wname, status))
        
    
    tb_no = int(input())
    
    a = findWaiterWiseTotalNoOdTables(t_list)
    for i in sorted(a.keys()):
        print("{}-{}".format(i,a[i]))
        
    b = findWaiterNameByTableNo(t_list, tb_no)
    if b == None:
        print("No Table Found")
    else:
        print(b)
    
    
```
